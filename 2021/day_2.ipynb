```python
aim = 0
horizontal = 0
depth = 0

for i in range(len(input)):
    coord = input[i].split(' ')[0]
    value = int(input[i].split(' ')[1])

    if coord == 'down':
        aim += value
    elif coord == 'up':
        aim -= value
    elif coord == 'forward':
        horizontal += value
        depth += aim * value

    print(f'Command #{i} {input[i]}: aim = {aim}, horizontal = {horizontal}, depth = {depth}')
```
